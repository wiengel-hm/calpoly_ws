### Concatenation & Stacking

Let’s say we have two 1×3 matrices:

$$
a = \begin{bmatrix} 1 & 2 & 3 \end{bmatrix}, \quad
b = \begin{bmatrix} 4 & 5 & 6 \end{bmatrix}
$$

---

| Operation | Code | Result |
|-----------|------|--------|
| **Concatenation** (1D result) | `np.concatenate([a, b])` | $\begin{bmatrix} 1 & 2 & 3 & 4 & 5 & 6 \end{bmatrix}$ |
| **Stacking** (2D result) | `np.stack([a, b])` | $\begin{bmatrix} \begin{bmatrix} 1 & 2 & 3 \end{bmatrix} \\ \begin{bmatrix} 4 & 5 & 6 \end{bmatrix} \end{bmatrix} \Rightarrow \text{shape }(2, 3)$ |

---

Now let’s say we have two **2×2** matrices:

$$
A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}, \quad
B = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}
$$

---

| Operation | Code | Result / Shape |
|-----------|------|----------------|
| **Concatenate along rows** (vertical) | `np.concatenate([A, B], axis=0)` | $\begin{bmatrix} 1 & 2 \\ 3 & 4 \\ 5 & 6 \\ 7 & 8 \end{bmatrix} \Rightarrow \text{shape }(4, 2)$ |
| **Concatenate along columns** (horizontal) | `np.concatenate([A, B], axis=1)` | $\begin{bmatrix} 1 & 2 & 5 & 6 \\ 3 & 4 & 7 & 8 \end{bmatrix} \Rightarrow \text{shape }(2, 4)$ |
